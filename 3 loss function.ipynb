```python
from sklearn.datasets import make_classification
 
X, y = make_classification(
    n_samples=100, # 1000 observations 
    n_features=2, # 5 total features
    n_informative=1, # 3 'useful' features
    n_redundant = 0,
    n_classes=2, # binary target/label 
    n_clusters_per_class = 1,
    hypercube = False,
    random_state=41,
    class_sep=15  # if you want the same results as mine
)
```

```python
X
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10,6))
plt.scatter(X[:,0],X[:,1],c = y,cmap = 'winter',s=40)
```

```python
def perceptron(X,y):
    
    w1 = w2 = b = 1
    lr = 0.1   #learning rate
    for j in range(100):
        for i in range(X.shape[0]):
            
            z= w1*X[i][0] + w2*X[i][1] +b
            if z*y[i] < 0:
                w1 = w1 +lr*y[i]*X[i][0]
                w2 =w2+lr*y[i]*X[i][1]
                b = b + lr*y[i]
                
    return w1,w2,b
```

```python
w1,w2,b = perceptron(X,y)
```

```python
m = -(w1/w2)
c = -(b/w2)

print(m,c)
```

```python
import numpy as np
x_input = np.linspace(-3,3,100)
y_input = m*x_input +c
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth =3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=50)
plt.ylim(-3,2)
```
